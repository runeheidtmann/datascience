```python
class Branche:
  def __init__(self, code, name,locked=True):
    self.code = code
    self.name = name
    self.locked = locked
    self.children = []


import csv
brancher = []

with open('csv_da.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    for row in spamreader:
        if row[2] == "1":
            print(row[1]+" - " +row[3])
            top = Branche(row[1],row[3])
            brancher.append(top)
        if row[2] == "2":
            print("   -"+row[1]+" - " +row[3])
            second = Branche(row[1],row[3])
            top.children.append(second)
        if row[2] == "3":
            print("      -"+row[1]+" - " +row[3])
            third = Branche(row[1],row[3])
            second.children.append(third)
        if row[2] == "4":
            print("          -"+row[1]+" - " +row[3])
            fourth = Branche(row[1],row[3])
            third.children.append(fourth)
        if row[2] == "5":
            print("              -"+row[1]+" - " +row[3])
            fifth = Branche(row[1],row[3])
            fourth.children.append(fifth)
           
```

```python
for branche in brancher:
    print(branche.name)
    for b in branche.children:
        print("   + "+b.name)
        if len(b.children) < 1:
            b.locked = False
            print("locked")
            
        for c in b.children:
            print("      + "+c.name)
            if len(c.children) < 1:
                c.locked = False
                print("locked")
            for d in c.children:
                print("          + "+d.name)
                if len(d.children) < 1:
                    d.locked = False
                    print("locked")
                for e in d.children:
                    print("              - "+e.name)
                    if len(e.children) < 1:
                        e.locked = False
                        print("locked")
                   
```

```python
import json


json_data = json.dumps(brancher, default=lambda o: o.__dict__)
print(json_data)
```
